```python
import pandas as pd

df_wt = pd.read_csv('epoch_10_5e-05_eval_results_with_test.csv')
styled_df = df_wt.style.set_table_styles([
    {'selector': 'tr:nth-of-type(even)', 'props': [('background-color', 'lightgrey'), ('color', 'black')]},
    {'selector': 'tr:nth-of-type(odd)', 'props': [('background-color', 'white'), ('color', 'black')]}
])
styled_df
```

```python
df_wt_wop = pd.read_csv('epoch_10_5e-05_eval_results_wo_pretrain_wtest.csv')
styled_df = df_wt_wop.style.set_table_styles([
    {'selector': 'tr:nth-of-type(even)', 'props': [('background-color', 'lightgrey'), ('color', 'black')]},
    {'selector': 'tr:nth-of-type(odd)', 'props': [('background-color', 'white'), ('color', 'black')]}
])
styled_df
```

```python
df = pd.concat([df_wt, df_wt_wop], ignore_index=True)
df = df.sort_values(by='test_f1', ascending=False)
df
```

```python
def print_pretty(df):
    styled_df = df.style.set_table_styles([
    {'selector': 'tr:nth-of-type(even)', 'props': [('background-color', 'lightgrey'), ('color', 'black')]},
    {'selector': 'tr:nth-of-type(odd)', 'props': [('background-color', 'white'), ('color', 'black')]}
])
    return styled_df
pretty_df = print_pretty(df)
pretty_df
```
